```python
# matplotlib으로 차트 그리기
# anscombe의 시각화 강조 주장을 검증
# 일반적인 가정 : 복수의 데이터 셋이 통계 수치, 상관관계, 회귀선이 같다면 비슷한 데이터 셋
# anscombe의 주장 : 데이터를 시각화 하지 않으면 데이터의 분포도를 알기 어렵다 -> 시각화 강조

# 데이터 셋 불러오가
import seaborn as sns
# 샘플 데이터 목록
print(sns.get_dataset_names())
#anscombe 데이터 셋 불러오기
anscombe = sns.load_dataset("anscombe")
anscombe
```

```python
# 데이터 세브셋의 갯수 확인
print(anscombe['dataset'].nunique())# 유니크한 값의 개수
```

```python
# 서브셋 분할
ds1 = anscombe[anscombe['dataset'] == 'I']  # 데이터 셋 1
ds2 = anscombe[anscombe['dataset'] == 'II'] # 데이터 셋 2
ds3 = anscombe[anscombe['dataset'] == 'III'] # 데이터 셋 3
ds4 = anscombe[anscombe['dataset'] == 'IV'] # 데이터 셋 4
```

```python
ds1.describe() # 데이터 셋 1의 통계량 확인
```

```python
ds2.describe() # 데이터 셋 2의 기초 통계량
```

```python
import matplotlib.pyplot as plt
# 데이터 셋 1 그려보기
plt.plot(ds1['x'], ds1['y'], 'o')
plt.plot(ds1['x'], ds2['y'], '*')
plt.plot(ds1['x'], ds3['y'], '^')
```

```python
# 데이터 셋 2의 그래프
plt.plot(ds2['x'], ds2['y'], 'v', color = "red")
```

```python
%matplotlib inline
# anscombe 4분할 그리기
# 그래프의 기본 틀
fig = plt.figure(figsize=(12, 12))
# 기본 틀에 격자 추가
axes1 = fig.add_subplot(2, 2, 1) # 2 x 2 격자의 1번 영역
axes2 = fig.add_subplot(2, 2, 2) # 2 x 2 격자의 2번 영역
axes3 = fig.add_subplot(2, 2, 3) # 2 x 2 격자의 3번 영역
axes4 = fig.add_subplot(2, 2, 4) # 2 x 2 격자의 4번 영역
# 격자에 그래프 그리기
axes1.plot(ds1['x'], ds1['y'], 'o')
axes2.plot(ds2['x'], ds2['y'], '+', color="yellow")
axes3.plot(ds3['x'], ds3['y'], '*', color='red')
axes4.plot(ds4['x'], ds4['y'], 'v', color='green')
# 서브 플랏에 타이틀 출력
axes1.set_title("DataSet 1")
axes2.set_title("DataSet 2")
axes3.set_title("DataSet 3")
axes4.set_title("DataSet 4")
# 기본 틀 타이틀 출력
fig.suptitle("Anscombe DataSets")
# 레이아웃 조정
```
