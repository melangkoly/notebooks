```python
# matplotlib으로 차트 그리기
# anscombe의 시각화 강조 주장을 검증
# 일반적인 가정 : 복수의 데이터 셋이 통계 수치, 상관관계, 회귀선이 같다면 비슷한 데이터 셋
# anscombe의 주장 : 데이터를 시각화 하지 않으면 데이터의 분포도를 알기 어렵다 -> 시각화 강조

# 데이터 셋 불러오가
import seaborn as sns
# 샘플 데이터 목록
print(sns.get_dataset_names())
#anscombe 데이터 셋 불러오기
anscombe = sns.load_dataset("anscombe")
anscombe
```

```python
# 데이터 세브셋의 갯수 확인
print(anscombe['dataset'].nunique())# 유니크한 값의 개수
```

```python
# 서브셋 분할
ds1 = anscombe[anscombe['dataset'] == 'I']  # 데이터 셋 1
ds2 = anscombe[anscombe['dataset'] == 'II'] # 데이터 셋 2
ds3 = anscombe[anscombe['dataset'] == 'III'] # 데이터 셋 3
ds4 = anscombe[anscombe['dataset'] == 'IV'] # 데이터 셋 4
```

```python
ds1.describe() # 데이터 셋 1의 통계량 확인
```

```python
import matplotlib.pyplot as plt
# 데이터 셋 1 그려보기
plt.plot(ds1['x'], ds1['y'], 'o')
plt.plot(ds1['x'], ds2['y'], '*')
plt.plot(ds1['x'], ds3['y'], '^')
```

```python
# 데이터 셋 2의 그래프
plt.plot(ds2['x'], ds2['y'], 'v', color = "red")
```

```python
# mtcars 데이터셋을 이용한 산점도 그래프 (Scatter plot)
import pandas as pd
mtcars = pd.read_csv("./data/mtcars.csv") # 헤더가 있고, 구분자가 (,)
mtcars
```

```python
from IPython.display import display
# EDA
print("Size:", mtcars.shape) # Record = 32, Column = 12
# 위쪽 데이터 확인
print(mtcars.head())
# 꼬리 데이터 확인
print(mtcars.tail())
```

```python
# 기초 통계량 확인
mtcars.describe()
```

```python
# mpg(연비) vs wt(중량) -> 산점도 그리기
plt.scatter(mtcars['mpg'], mtcars['wt'], marker='o')
plt.xlabel('Weight')
plt.xlabel('Mile per Gallon')
plt.title("Weight vs MPG")

plt.show() # 그리기
```

```python
# 상관 계수
print("상관계수(wt vs mpg)")
mtcars.corr() # 상관계수 - 모든 수치 변수들 사이의 상관계수
mtcars['wt'].corr(mtcars['mpg']) # wt 변수와 mpg 변수 사이의 상관 계수 -> 역상관
```

```python
# 상자 그림(Box Plot)
# plt.boxplot(mtcars['mpg']) : 단일 변수의 상자 그림
plt.boxplot([mtcars['mpg'], mtcars['wt']])
plt.show()

print("mpg 기초 통계량")
display(mtcars['mpg'].describe())
```

```python
# 분위 나누깃
quantiles = mtcars.quantile([0.25, 0.5, 0.75]) # 모든 변수의 분위수
display(quantiles)
```

```python
# mpg의 1, 3 사분위 수
first_quartile = mtcars['mpg'].quantile(0.25) # 1사분위
third_quartile = mtcars['mpg'].quantile(0.75) # 3사분위
print("1사분위수 {}. 3사분위수 {}".format(first_quartile, third_quartile))
# IQR(inter Quartile Range) : Q3 - Q1
irq = third_quartile - first_quartile
print("IQR of mpg: {}".format(irq))
```

```python
# mpg의 상단 극단치 경계
# Q3 + IQR * 1.5 
border_top = third_quartile + irq * 1.5
print("상단 극단치 경계: ", border_top)
border_bottom = first_quartile - irq * 1.5
print("하단 극단치 경계: ", border_bottom)
```

```python
# mpg가 극단치를 가진 자동차?
# mpg가 상단 극단치를 초과한 레코드
# mtcars['mpg'] > border_top Boolean 연산
outliers = mtcars[mtcars['mpg'] > border_top]
print("Outliers : ")
```
